```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.family'] = 'Times New Roman'

video = [0.648,0.575,0.647]
audio = [0.533,0.53,0.537]
chat = [0.347, 0.346, 0.344]
multimodal = [0.718, 0.722, 0.723] 

year = ['64', '128', '256']

df = pd.DataFrame({'video' : video, 'audio' : audio, 'chat' : chat, 'multimodal':multimodal}, index = year)


# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(10,5))
bar_width = 0.1

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(3)

labels=['LSTM('+r'$V_{score}$'+')','LSTM('+r'$A_{mfcc}$'+')','LSTM('+r'$C_{label}$'+')','LSTM('+r'$V_{score}$'+') + LSTM('+r'$A_{mfcc}$'+') + LSTM('+r'$C_{label}$'+')']

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(0.5+index, df['video'], bar_width, alpha=0.4, color='red', label=labels[0])

b2 = plt.bar(0.5+index + bar_width, df['audio'], bar_width, alpha=0.4, color='blue', label=labels[1])

b3 = plt.bar(0.5+index + 2 * bar_width, df['chat'], bar_width, alpha=0.4, color='green', label=labels[2])

b3 = plt.bar(0.5+index + 3 * bar_width, df['multimodal'], bar_width, alpha=0.4, color='violet', label=labels[3])

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(0.5+bar_width, 0.5+3 + bar_width, 1), year)

# x축, y축 이름 및 범례 설정
#plt.xlabel('Features', size = 13)
plt.ylabel('F1-score', size = 13)
plt.legend(bbox_to_anchor=(0.17, 1),ncol=2)#,prop = dict(style='italic')
plt.show()
```
